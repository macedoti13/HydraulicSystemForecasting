```python
import pandas as pd 
```

```python
df = pd.read_parquet("data/original_data/water_consumption.parquet")
df.head()
```

```python
df["GMB 1 (10 OFF/ 90 ON)"] = df["GMB 1 (10 OFF/ 90 ON)"].replace(10, 0).replace(90, 1)
df["GMB 2(10 OFF/ 90 ON)"]  = df["GMB 2(10 OFF/ 90 ON)"].replace(10, 0).replace(90, 1)
df.rename(columns={
    "GMB 1 (10 OFF/ 90 ON)": "gmb_1_is_on", 
    "GMB 2(10 OFF/ 90 ON)": "gmb_2_is_on", 
    "PRESSÃO (mca)": "pressure_(mca)",
    "VAZÃO ENTRADA (L/S)" :"flow_in_(l/s)",
    "NÍVEL RESERVATÓRIO (%)": "reservoir_level_(%)",
    "DATA/HORA": "timestamp",
    }, inplace=True)

df.head()
```

```python
df.to_parquet("data/preprocessed_data/water_consumption_cleaned_0.parquet")
```
